## Sulfur isotope balance
- R$_{system}$X$_{system}$ = R$_{MI, meas}$X$_{MI, meas}$C$_{MI, meas}$ + R$_{sulf}$X$_{sulf}$C$_{sulf}$ + R$_{Ol}$X$_{Ol}$C$_{Ol}$
- R represents the ratio of 34/32S in each phase
- X represents the mass balance of each phase 
- C represents the concentration of S in each phase

We can simply this by saying that X$_{system}$ = 1 (e.g. initial mass of system is 1), and that there is no S in the olivine
-  R$_{system}$ = R$_{MI, meas}$X$_{MI, meas}$C$_{MI, meas}$ + R$_{sulf}$X$_{sulf}$C$_{sulf}$

We can now substitude alpha, the fractionatoin factor between the sulfide and the measured melt
- $\alpha$$_{sulf/melt}$ = R$_{sulf}$/R$_{MI, meas}$
- R$_{sulf}$=  R$_{MI, meas}$ $\alpha$$_{sulf/melt}$ 

Which converts the above equation to: 
- R$_{system}$ = R$_{MI, meas}$X$_{MI, meas}$C$_{MI, meas}$ + R$_{MI, meas}$ $\alpha$$_{sulf/melt}$X$_{sulf}$C$_{sulf}$

We can also substitute X$_{MI, meas}$ in terms of the amount of PEC
- X$_{MI, meas}$=1/(1+PEC/100)

Which takes this to:

- R$_{system}$ =$\frac{R_{MI, meas}} {(1+PEC/100)}$ C$_{MI, meas}$ + R$_{MI, meas}$ $\alpha$$_{sulf/melt}$X$_{sulf}$C$_{sulf}$

We then guess R system from the least fractionated isotopes. We use a linear vector of PEC, we can guess the S content of the sulfide, and the S content of the glass is got from a regressoin wtih PEc. We are trying to re-arrange for the resulting S isotope composition of the measured melt. 

- R$_{MI, meas}$=$\frac{R_{system}}{  \frac{C_{MI, meas}} {(1+PEC/100)}  + \alpha_{sulf/melt}X_{sulf}C_{sulf}  }$

```python
PEC_lin=np.linspace(0, 60, 100)
X_melt_loop=1/(1+(PEC_lin/100))
```

```python
def S_isotope_frac_model(alpha=None, PEC_lin=None, S_Sulf=32*10**4):

    # Amount of melt remaining based on how petrolog defines mass balance
    X_melt_loop=1/(1+(PEC_lin/100))
    
    
    # Regression linking S in the melt to the amount of PEC for these samples. This is the 
    # C in the MI term
    Sulf_melt_lin=stats_PEC_S['Grad']*PEC_lin+stats_PEC_S['Int']
    
    # Cumulative mass of S in the sulfide (following Kiseeva and Wood, 2015)
    X_Sulf_lin=(Sulf_melt_lin[0]-(X_melt_loop*Sulf_melt_lin))/S_Sulf
    
    # Setting up some empty variables to fill in in the loop
    d34_melt_loop=np.empty(len(PEC_lin), float)
    S3432_melt_loop=np.empty(len(PEC_lin), float)
    S3432_sulf_loop=np.empty(len(PEC_lin), float)
    # Why would this be inverted?
    alpha_34S=1/alpha
    init_3432S=ss.convert_d34_to_3432S(init_34dS)
    print('init 3432S')
    print(init_3432S)

    for i in range(0, len(PEC_lin)):
        
        # Set at the start, as always re-equilibrate
        # This is the overal ratio of the system * concentratoin - so 34S total. 
        constant=init_3432S*Sulf_melt_lin[0]
        
        if i==0:
            # d34S in the melt is what we set it to at the start
            S3432_melt_loop[i]=init_3432S
            # d34S in the sulfide is what we set it in the melt * alpha. 
            S3432_sulf_loop[i]=S3432_melt_loop[i]*alpha_34S

        else:
            
            S3432_melt_loop[i]=(constant)/(Sulf_melt_lin[i]*X_melt_loop[i]+alpha_34S*X_Sulf_lin[i]*S_Sulf)
            S3432_sulf_loop[i]=S3432_melt_loop[i]*alpha_34S

    d34_melt_loop=ss.convert_3432S_to_d34(S3432_melt_loop)
    d34_sulf_loop=ss.convert_3432S_to_d34(S3432_sulf_loop)
    return d34_melt_loop, d34_sulf_loop, Sulf_melt_lin, S3432_melt_loop, S3432_sulf_loop
```

```python
d34_melt_loop_997, d34_sulf_loop_997, Sulf_melt_lin,  S3432_melt_loop, S3432_sulf_loop=S_isotope_frac_model(alpha=0.997, PEC_lin=PEC_lin)
```
